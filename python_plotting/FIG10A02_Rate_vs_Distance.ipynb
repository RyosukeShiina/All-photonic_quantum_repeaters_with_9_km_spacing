```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter
import csv
import pandas as pd
print(pd.__version__)
print(np.__version__)
```

```python
plt.rcParams['font.family'] = 'Times New Roman' 
plt.rcParams['mathtext.fontset'] = 'stix' 
plt.rcParams["font.size"] = 12
plt.rcParams['xtick.labelsize'] = 17
plt.rcParams['ytick.labelsize'] = 17 

plt.rcParams['xtick.direction'] = 'in' 
plt.rcParams['ytick.direction'] = 'in' 
plt.rcParams['axes.grid'] = True 

plt.rcParams["xtick.minor.visible"] = True  
plt.rcParams["ytick.minor.visible"] = True  
plt.rcParams['xtick.top'] = True  
plt.rcParams['ytick.right'] = True  

plt.rcParams["legend.fancybox"] = False 
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["legend.edgecolor"] = 'black'
plt.rcParams["legend.markerscale"] = 5
```

```python
def main():
    xlist2 = []
    for i in range(1500):
        xlist2.append(2*(i+1))

    xlist3 = []
    for i in range(1000):
        xlist3.append(3*(i+1))
    
    xlist3p3 = []
    for i in range(1000):
        xlist3p3.append(3.3*(i+1))
    
    xlist4 = []
    for i in range(750):
        xlist4.append(4*(i+1))

    xlist10 = []
    for i in range(300):
        xlist10.append(10*(i+1))

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW3_2km_15.csv', header=None)
    Rate_vs_Distance_ML_UW3_k15_2km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW3_k15_2km = [x*15 for x in Rate_vs_Distance_ML_UW3_k15_2km]
    
    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW3_3km_15.csv', header=None)
    Rate_vs_Distance_ML_UW3_k15_3km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW3_k15_3km = [x*15 for x in Rate_vs_Distance_ML_UW3_k15_3km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW3_3p3km_15.csv', header=None)
    Rate_vs_Distance_ML_UW3_k15_3p3km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW3_k15_3p3km = [x*15 for x in Rate_vs_Distance_ML_UW3_k15_3p3km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW3_4km_15.csv', header=None)
    Rate_vs_Distance_ML_UW3_k15_4km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW3_k15_4km = [x*15 for x in Rate_vs_Distance_ML_UW3_k15_4km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW2_2km_15.csv', header=None)
    Rate_vs_Distance_ML_UW2_k15_2km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW2_k15_2km = [x*15 for x in Rate_vs_Distance_ML_UW2_k15_2km]
    
    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW2_3km_15.csv', header=None)
    Rate_vs_Distance_ML_UW2_k15_3km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW2_k15_3km = [x*15 for x in Rate_vs_Distance_ML_UW2_k15_3km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW2_3p3km_15.csv', header=None)
    Rate_vs_Distance_ML_UW2_k15_3p3km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW2_k15_3p3km = [x*15 for x in Rate_vs_Distance_ML_UW2_k15_3p3km]

    df = pd.read_csv('FIG_Data/Rate_vs_Distance_ML_and_LP/Rate_vs_Distance_ML_UW2_4km_15.csv', header=None)
    Rate_vs_Distance_ML_UW2_k15_4km = df.values.flatten().tolist()
    Total_Rate_vs_Distance_ML_UW2_k15_4km = [x*15 for x in Rate_vs_Distance_ML_UW2_k15_4km]

    #PLOB bound
    eta_list = [10**(-L/22) for L in xlist10]
    PLOB_rate = [-np.log2(1 - eta) if eta < 1 else 0 for eta in eta_list]
    
    fig = plt.figure(dpi=1200)
    fig_1 = fig.add_subplot(111)
    fig_1.plot(xlist2, Total_Rate_vs_Distance_ML_UW3_k15_2km, label="$k=15$, $L_0=2km$ (UW3)", color="orange")
    fig_1.plot(xlist3, Total_Rate_vs_Distance_ML_UW3_k15_3km, label="$k=15$, $L_0=3km$ (UW3)", color="#e377c2")
    fig_1.plot(xlist3p3, Total_Rate_vs_Distance_ML_UW3_k15_3p3km, label="$k=15$, $L_0=3.3km$ (UW3)", color="r")
    fig_1.plot(xlist4, Total_Rate_vs_Distance_ML_UW3_k15_4km, label="$k=15$, $L_0=4km$ (UW3)", color="g")
    
    fig_1.plot(xlist10, PLOB_rate, label="PLOB Bound", color="blue", linestyle="--")

    fig_1.plot(xlist2, Total_Rate_vs_Distance_ML_UW2_k15_2km, label="$k=15$, $L_0=2km$ (UW2)", color="orange", linestyle="--", zorder=4, linewidth="2.5")
    fig_1.plot(xlist3, Total_Rate_vs_Distance_ML_UW2_k15_3km, label="$k=15$, $L_0=3km$ (UW2)", color="#e377c2", linestyle="--", zorder=4, linewidth="2.5")
    fig_1.plot(xlist3p3, Total_Rate_vs_Distance_ML_UW2_k15_3p3km, label="$k=15$, $L_0=3.3km$ (UW2)", color="r", linestyle="--", zorder=4, linewidth="2.5")
    fig_1.plot(xlist4, Total_Rate_vs_Distance_ML_UW2_k15_4km, label="$k=15$, $L_0=4km$ (UW2)", color="g", linestyle="--", zorder=4, linewidth="2.5")

    fig_1.set_xlabel(r"$L_{\mathrm{total}}$ [km]")
    fig_1.set_ylabel(r"$R_{\mathrm{protocol\ run}}$ [ebit/protocol run]")
    fig_1.legend(ncol=2, bbox_to_anchor=(0, 1.1, 1., 0.102), loc=3)
    fig_1.grid(which='minor', alpha=0.2)
    fig_1.grid(which='major', alpha=0.5)
    fig_1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    #fig_1.ticklabel_format(style="sci",  axis="y",scilimits=(0,0))
    plt.xlim(0, 3000)
    plt.ylim(1e-4, 20)
    plt.yscale("log")
    plt.show()

    fig.savefig('FIG10A02.png', bbox_inches="tight", pad_inches=0.05) 

if __name__ == "__main__":
    main()
```
